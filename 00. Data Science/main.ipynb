#### ***NumPy (as np)***
> NumPy is a Python library that provides a simple yet powerful data structure: the n-dimensional array. This is the foundation on which almost all the power of Python’s data science toolkit is built, and learning NumPy is the first step on any Python data scientist’s journey.

#### ***Matplotlib.pyplot (as plt)***
> Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. Matplotlib makes easy things easy and hard things possible.
matplotlib.pyplot is a collection of functions that make matplotlib work like MATLAB. Each pyplot function makes some change to a figure: e.g., creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc.



```python
# learning numpy
import numpy as np
```

```python
CURVE_CENTER = 80

grades = np.array([72, 35, 64, 88, 51, 90, 74,  12])
def curve(grades):
    average = grades.mean()
    change = CURVE_CENTER - average
    new_grade = grades + change
    return np.clip(new_grade, grades, 100)

curved_grades = curve(grades)
curved_grades
```

```python
# learning matplotlib
import matplotlib.pyplot as plt
```

```python
# taking numpy values to matplotlib

# reference: 
# ORIGINAL GRADES: [72, 35, 64, 88, 51, 90, 74,  12]
# CURVED GRADES: [91.25,  54.25,  83.25, 100.  ,  70.25, 100.  ,  93.25,  31.25]

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.bar(range(len(grades)), grades, color='blue', width=0.7)
plt.title('Original Grades')
plt.xlabel('Student')
plt.ylabel('Grade')

plt.subplot(1,2,2)
plt.bar(range(len(curved_grades)), curved_grades, color='green', width=0.7)
plt.title('Curved Grades')
plt.xlabel('Student')
plt.ylabel('Curved Grade')

plt.tight_layout()
plt.show()
```

## Learning with basic projects (numpy)

***1. Vectorized Operations:***
Explore the power of vectorized operations in NumPy. Perform basic arithmetic operations on arrays and compare the speed with traditional loops. Understand the concept of broadcasting.

```python
# Vectorized Operation with numpy

arr1 = [82, 18, 87, 35, 49]
arr2 = [18, 30, 29, 78, 51]

arr_add = arr1 + arr2

arr_add_sorted = sorted(arr_add)

print("Added Array:", arr_add)
print("Sorted Array:", arr_add_sorted)

add_each = [a + b for a, b in zip(arr1, arr2)]
print(add_each)
print(sorted(add_each))
```

## ***Axes***
>In Matplotlib, ax is an instance of the Axes class. The Axes class is a part of the Matplotlib library and represents the set of plotting axes (i.e., the region of the image with the data space). In simple terms, it's the area where you create your plots.

> When you create a figure using plt.subplots(), you get both a Figure instance and an array of Axes instances. If you're creating a single plot, there will be just one set of axes. If you're creating a grid of plots, ax would be an array containing all the individual plots.

```python
fig, ax = plt.subplots()

# Plot arr1
ax.plot(arr1, label='arr1', marker='o')

# Plot arr2
ax.plot(arr2, label='arr2', marker='o')

plt.xlabel('arr1')
plt.ylabel('arr2')

# Show legend
ax.legend()

# Show the plot
plt.show()
```
